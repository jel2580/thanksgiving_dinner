```python
def extract_age(string):
    if pd.isnull(string):
        return None
    else:
        x=string.split()
        y=x[0].replace("+","")
        z=int(y)
        return z
```

```python
extract_age("29 - 35")
```

```python
data['int_age']=data['Age'].apply(extract_age)
```

```python
data['int_age']
```

```python
data['int_age'].describe()
```

```python
data['How much total combined money did all members of your HOUSEHOLD earn last year?'].value_counts()
```

```python
def extract_income(string):
    if pd.isnull(string):
        return None
    else:
        x=string.split()[0]
        if x=="Prefer":
            return None
        else:
            y=x.replace("$","")
            y=y.replace(",","")
            return int(y)
```

```python
extract_income("$40,000 - $80,000")
```

```python
data['int_income']=data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(extract_income)
```

```python
data['int_income'].describe()
```

```python
lower_income=data[data['int_income']<150000]
lower_income['How far will you travel for Thanksgiving?'].value_counts()
```

```python
higher_income=data[data['int_income']>150000]
higher_income['How far will you travel for Thanksgiving?'].value_counts()
```
